```python
import pandas as pd
spotify = pd.read_csv('/content/drive/MyDrive/Spotify.csv')
spotify
```

```python
canciones_pegadas =  spotify[(spotify['Region']=='global') & (spotify['Position']==1)]['Track Name'].unique()
canciones_pegadas
```

```python
spotify = spotify[spotify['Region'].isin(['global','ar'])]
spotify = spotify[spotify['Track Name'].isin(canciones_pegadas)]
spotify
```

```python
spotify['Track Name'] = spotify['Track Name'].replace({'Despacito (Featuring Daddy Yankee)':'Despacito'})
```

```python
spotifyN1 = spotify[spotify['Position']==1]
spotifyN1 = spotifyN1[['Track Name','Date','Region']]
spotifyN1
```

```python
import matplotlib.pyplot as plt #Pylplot te permite hacer cosas sencillas
import seaborn as sns #Seaborn ter permite hacer cosas mas chetas
```

```python
 orden = spotifyN1[spotifyN1['Region']=='global']['Track Name'].value_counts().index
 orden
```

```python
plt.figure(dpi = 150)
sns.countplot(data = spotifyN1,x='Track Name',palette=['#75AADB','#1DB954'],hue='Region', order=orden)
plt.xticks(rotation=90)
plt.title('Que tanto influye el ranking de argentina en el ranking global')
plt.ylabel('Numero de días en la posicion N1')
plt.xlabel('Nombre de la canción')
```
